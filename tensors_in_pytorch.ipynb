## Mathemetical Operations

#### Scalar Operations

```python
x = torch.rand(2,2)
x
```

```python
### add
x + 2

## subs
x - 3

## multi
x * 100

## division
x / 5

## int division
(x*100)//2

## mod
(x*100)//2 % 2

### power
x ** 2
```

#### Element wise operation

```python
a = torch.rand(3,3)
b = torch.rand(3,3)
print(a)
print(b)
```

```python
a + b

#sub
a-b

#multi
a*b

#div
a/b
```

```python
c = torch.tensor([1,-2,-3,4])
```

```python
## neg
torch.neg(c)
```

```python
d = torch.tensor([7.9,12.1,13.10,55.01])
torch.round(d)
```

#### Reduction Operations

```python
e = torch.randint(size=(2,3),low=0,high=10,dtype=torch.float32)
e
```

```python
torch.sum(e)
```

```python
torch.sum(e,dim=0) ## sum of column
```

```python
torch.sum(e,dim=1) ## sum of row
```

```python
## mean
torch.mean(e)

## row wise
torch.mean(e,dim=1)
```

```python
## median
torch.median(e)
```

```python
torch.max(e)
```

```python
torch.min(e)
```

```python
 torch.prod(e)
```

```python
torch.std(e)
```

```python
torch.var(e)
```

```python
## return the position of biggest element
torch.argmax(e)
```

```python
## return the position of biggest element
torch.argmin(e)
```

#### MATRIX OPERATIONS

```python
f = torch.randint(size=(2,3),low=0,high=10)
g = torch.randint(size=(3,2),low=0,high=10)

print(f)
print(g)
```

```python
torch.matmul(f,g)
```

```python
## dot product
vec1 = torch.tensor([1,2,3])
vec2 = torch.tensor([4,5,6])

torch.dot(vec1,vec2)
```

```python
## transpose
torch.transpose(f,0,1)
```
